## General model analysis - correlation matrices

### Subtask:
Calculate and display the correlation matrix for each species in the General Model table, calculate and display the overall correlation matrix for the General Model table, and save these correlation matrices as images.


**Reasoning**:
I need to calculate and display the correlation matrices for each species and overall in the general model table, then save them as images.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Re-load the original df_general to access the 'Species' column
df_general_original = pd.read_csv("/content/General_Model_Table_clusters.csv")

# Get unique species
unique_species = df_general_original['Species'].unique()

# Drop unnecessary columns from the original dataframe for correlation calculation
df_general_corr = df_general_original.drop(columns=['Species', 'Dataset', 'File Name', 'clusters'])

# Function to calculate and plot correlation matrix
def plot_and_save_correlation_matrix(df, title, filename):
    """
    Calculates, plots, and saves the correlation matrix for a DataFrame
    with categorical 'S', 'I', 'R' values.
    """
    # Convert categorical data to numerical representation for correlation
    # Map 'S' to 0, 'I' to 1, 'R' to 2, and NaN to NaN
    df_numeric = df.apply(lambda x: x.map({'S': 0, 'I': 1, 'R': 2}))
    correlation_matrix = df_numeric.corr()

    plt.figure(figsize=(10, 8))
    # Changed cmap to 'viridis' for a prettier color scheme
    sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
    plt.title(title)
    plt.savefig(filename)
    plt.show()
    display(correlation_matrix)

# Calculate and display correlation matrix for each species
print("--- General Model Analysis - Correlation Matrices per Species ---")
for species in unique_species:
    df_species = df_general_original[df_general_original['Species'] == species].drop(columns=['Species', 'Dataset', 'File Name', 'clusters'])
    if not df_species.empty:
        plot_and_save_correlation_matrix(df_species, f'General Model: Correlation Matrix for {species}', f'general_model_{species.replace(" ", "_")}_correlation_matrix.png')
    else:
        print(f"No data for species: {species}")

# Calculate and display overall correlation matrix
print("\n--- General Model Analysis - Overall Correlation Matrix ---")
plot_and_save_correlation_matrix(df_general_corr, 'General Model: Overall Correlation Matrix', 'general_model_overall_correlation_matrix.png')
```

## General model analysis - null percentages

### Subtask:
Calculate and display the percentage of nulls for each drug (binary label) per species in the General Model table, calculate and display the overall percentage of nulls for each drug in the General Model table, and save these results as CSV files.


**Reasoning**:
Reload the original General Model table to include the 'Species' column, define a function to calculate null percentages per species, iterate through species to apply the function, combine the results, calculate overall null percentages, display both results, and save them to CSV files.



```python
# 1. Reload the original 'General_Model_Table_clusters.csv'
df_general_original = pd.read_csv("/content/General_Model_Table_clusters.csv")

# 2. Define a function to calculate null percentages per species
def calculate_null_percentages_per_species(df, species_name):
    """Calculates null percentages for Tetracycline and Gentamicin for a given species."""
    df_species = df[df['Species'] == species_name]
    null_percentages = df_species[['Tetracycline', 'Gentamicin']].isnull().mean() * 100
    return null_percentages

# 3. Iterate through the unique species and get null percentages
unique_species = df_general_original['Species'].unique()
null_percentages_list = []

print("--- General Model Analysis - Null Percentages per Species ---")
for species in unique_species:
    null_percentages = calculate_null_percentages_per_species(df_general_original, species)
    null_percentages.name = species  # Name the series with the species name
    null_percentages_list.append(null_percentages)
    print(f"\nNull Percentages for {species}:")
    display(null_percentages)


# 4. Combine the null percentages for each species into a single DataFrame
df_null_percentages_per_species = pd.concat(null_percentages_list, axis=1).T

# 5. Calculate the overall percentage of nulls
overall_null_percentages = df_general_original[['Tetracycline', 'Gentamicin']].isnull().mean() * 100

# 6. Display the DataFrame containing null percentages per species
print("\n--- General Model Analysis - Combined Null Percentages per Species ---")
display(df_null_percentages_per_species)

# 7. Display the overall null percentages
print("\n--- General Model Analysis - Overall Null Percentages ---")
display(overall_null_percentages)

# 8. Save the DataFrame containing null percentages per species to a CSV file
df_null_percentages_per_species.to_csv('general_model_null_percentages_per_species.csv')

# 9. Save the overall null percentages to a CSV file
overall_null_percentages.to_csv('general_model_overall_null_percentages.csv')
```

## General model analysis - label balance

### Subtask:
Calculate and display the balance of labels for both drugs for each species in the General Model table, calculate and display the overall balance of labels for both drugs in the General Model table, and save these results as CSV files.


**Reasoning**:
Reload the original dataframe to include the 'Species' column, define a function to calculate label balances per species, iterate through unique species to calculate balances, concatenate results, calculate overall balances, display results, and save to CSV files.



```python
# 1. Reload the original dataframe from /content/General_Model_Table_clusters.csv to include the 'Species' column.
df_general_original = pd.read_csv("/content/General_Model_Table_clusters.csv")

# 2. Define a function that takes a DataFrame and a species name as input, calculates the (R+I)/S ratio for 'Tetracycline' and 'Gentamicin' for that species, and returns a DataFrame with these ratios.
def calculate_label_ratio_per_species(df, species_name):
    """Calculates the (R+I)/S ratio for Tetracycline and Gentamicin for a given species."""
    df_species = df[df['Species'] == species_name]

    tetracycline_counts = df_species['Tetracycline'].value_counts()
    gentamicin_counts = df_species['Gentamicin'].value_counts()

    tetracycline_ratio = (tetracycline_counts.get('R', 0) + tetracycline_counts.get('I', 0)) / tetracycline_counts.get('S', 0) if tetracycline_counts.get('S', 0) > 0 else float('inf')
    gentamicin_ratio = (gentamicin_counts.get('R', 0) + gentamicin_counts.get('I', 0)) / gentamicin_counts.get('S', 0) if gentamicin_counts.get('S', 0) > 0 else float('inf')

    return pd.DataFrame({'Tetracycline (R+I)/S Ratio': [tetracycline_ratio],
                         'Gentamicin (R+I)/S Ratio': [gentamicin_ratio]},
                        index=[species_name])

# 3. Iterate through the unique species in the original DataFrame, calling the function defined in the previous step for each species. Store the results for each species.
unique_species = df_general_original['Species'].unique()
label_ratio_list = []

print("--- General Model Analysis - Label (R+I)/S Ratio per Species ---")
for species in unique_species:
    label_ratio = calculate_label_ratio_per_species(df_general_original, species)
    label_ratio_list.append(label_ratio)
    print(f"\nLabel (R+I)/S Ratio for {species}:")
    display(label_ratio)


# 4. Concatenate the results for all species into a single DataFrame.
df_label_ratio_per_species = pd.concat(label_ratio_list)

# 5. Calculate the overall (R+I)/S ratio for 'Tetracycline' and 'Gentamicin' across all species in the original DataFrame.
overall_tetracycline_counts = df_general_original['Tetracycline'].value_counts()
overall_gentamicin_counts = df_general_original['Gentamicin'].value_counts()

overall_tetracycline_ratio = (overall_tetracycline_counts.get('R', 0) + overall_tetracycline_counts.get('I', 0)) / overall_tetracycline_counts.get('S', 0) if overall_tetracycline_counts.get('S', 0) > 0 else float('inf')
overall_gentamicin_ratio = (overall_gentamicin_counts.get('R', 0) + overall_gentamicin_counts.get('I', 0)) / overall_gentamicin_counts.get('S', 0) if overall_gentamicin_counts.get('S', 0) > 0 else float('inf')

overall_label_ratio = pd.DataFrame({'Tetracycline (R+I)/S Ratio': [overall_tetracycline_ratio],
                                     'Gentamicin (R+I)/S Ratio': [overall_gentamicin_ratio]},
                                    index=['Overall'])


# 6. Display the DataFrame containing label ratios per species.
print("\n--- General Model Analysis - Combined Label (R+I)/S Ratio per Species ---")
display(df_label_ratio_per_species)

# 7. Display the Series containing the overall label ratios.
print("\n--- General Model Analysis - Overall Label (R+I)/S Ratio ---")
display(overall_label_ratio)

# 8. Save the DataFrame containing label ratios per species to a CSV file named general_model_label_ratio_per_species.csv.
df_label_ratio_per_species.to_csv('general_model_label_ratio_per_species.csv')

# 9. Save the Series containing overall label ratios to a CSV file named general_model_overall_label_ratio.csv.
overall_label_ratio.to_csv('general_model_overall_label_ratio.csv')
```

Calculate and display the balance per drug for the Salmonella only table, and save this result as a CSV file.



```python
# 1. Calculate the value counts for each drug in the df_salmonella DataFrame.
salmonella_label_counts = df_salmonella.apply(pd.Series.value_counts)

# 2. Calculate the (R+I)/S ratio for each drug.
salmonella_label_ratio = pd.DataFrame(index=['(R+I)/S Ratio'])
for drug in salmonella_label_counts.columns:
    r_count = salmonella_label_counts.get(drug, {}).get('R', 0)
    i_count = salmonella_label_counts.get(drug, {}).get('I', 0)
    s_count = salmonella_label_counts.get(drug, {}).get('S', 0)

    ratio = (r_count + i_count) / s_count if s_count > 0 else float('inf')
    salmonella_label_ratio[drug] = [ratio]


# 3. Display the label ratio per drug.
print("--- Salmonella Only Analysis - Label (R+I)/S Ratio per Drug ---")
display(salmonella_label_ratio)

# 4. Save the label ratio per drug to a CSV file named 'salmonella_only_label_ratio_per_drug.csv'.
salmonella_label_ratio.to_csv('salmonella_only_label_ratio_per_drug.csv')
```

```python
# 1. Reload the original dataframe from /content/General_Model_Table_clusters.csv to include the 'Species' column.
df_general_original = pd.read_csv("/content/General_Model_Table_clusters.csv")

# 2. Define a function that takes a DataFrame and a species name as input, calculates the (R+I)/S ratio for 'Tetracycline' and 'Gentamicin' for that species, and returns a DataFrame with these ratios.
def calculate_label_ratio_per_species(df, species_name):
    """Calculates the (R+I)/S ratio for Tetracycline and Gentamicin for a given species."""
    df_species = df[df['Species'] == species_name]

    tetracycline_counts = df_species['Tetracycline'].value_counts()
    gentamicin_counts = df_species['Gentamicin'].value_counts()

    tetracycline_ratio = (tetracycline_counts.get('R', 0) + tetracycline_counts.get('I', 0)) / tetracycline_counts.get('S', 0) if tetracycline_counts.get('S', 0) > 0 else float('inf')
    gentamicin_ratio = (gentamicin_counts.get('R', 0) + gentamicin_counts.get('I', 0)) / gentamicin_counts.get('S', 0) if gentamicin_counts.get('S', 0) > 0 else float('inf')

    return pd.DataFrame({'Tetracycline (R+I)/S Ratio': [tetracycline_ratio],
                         'Gentamicin (R+I)/S Ratio': [gentamicin_ratio]},
                        index=[species_name])

# 3. Iterate through the unique species in the original DataFrame, calling the function defined in the previous step for each species. Store the results for each species.
unique_species = df_general_original['Species'].unique()
label_ratio_list = []

print("--- General Model Analysis - Label (R+I)/S Ratio per Species ---")
for species in unique_species:
    label_ratio = calculate_label_ratio_per_species(df_general_original, species)
    label_ratio_list.append(label_ratio)
    print(f"\nLabel (R+I)/S Ratio for {species}:")
    display(label_ratio)


# 4. Concatenate the results for all species into a single DataFrame.
df_label_ratio_per_species = pd.concat(label_ratio_list)

# 5. Calculate the overall (R+I)/S ratio for 'Tetracycline' and 'Gentamicin' across all species in the original DataFrame.
overall_tetracycline_counts = df_general_original['Tetracycline'].value_counts()
overall_gentamicin_counts = df_general_original['Gentamicin'].value_counts()

overall_tetracycline_ratio = (overall_tetracycline_counts.get('R', 0) + overall_tetracycline_counts.get('I', 0)) / overall_tetracycline_counts.get('S', 0) if overall_tetracycline_counts.get('S', 0) > 0 else float('inf')
overall_gentamicin_ratio = (overall_gentamicin_counts.get('R', 0) + overall_gentamicin_counts.get('I', 0)) / overall_gentamicin_counts.get('S', 0) if overall_gentamicin_counts.get('S', 0) > 0 else float('inf')

overall_label_ratio = pd.DataFrame({'Tetracycline (R+I)/S Ratio': [overall_tetracycline_ratio],
                                     'Gentamicin (R+I)/S Ratio': [overall_gentamicin_ratio]},
                                    index=['Overall'])


# 6. Display the DataFrame containing label ratios per species.
print("\n--- General Model Analysis - Combined Label (R+I)/S Ratio per Species ---")
display(df_label_ratio_per_species)

# 7. Display the Series containing the overall label ratios.
print("\n--- General Model Analysis - Overall Label (R+I)/S Ratio ---")
display(overall_label_ratio)

# 8. Save the DataFrame containing label ratios per species to a CSV file named general_model_label_ratio_per_species.csv.
df_label_ratio_per_species.to_csv(os.path.join('eda_results', 'general_model_label_ratio_per_species.csv')) # Save to eda_results directory

# 9. Save the Series containing overall label ratios to a CSV file named general_model_overall_label_ratio.csv.
overall_label_ratio.to_csv(os.path.join('eda_results', 'general_model_overall_label_ratio.csv')) # Save to eda_results directory
```

## Prepare data for evaluation (general model)

### Subtask:
Select the relevant drug columns from the merged DataFrame for the `General Model` dataset and convert 'S' and 'R' labels to numerical values. Handle missing values.


**Reasoning**:
Identify and process the original and predicted drug columns, convert labels to numerical values, and handle missing data for the general dataset.



```python
original_cols_general = [col for col in df_merged_general.columns if col.endswith('_original') and col not in ['File Name_original', 'Species_original', 'Dataset_original']]
prediction_cols_general = [col for col in df_merged_general.columns if col.endswith('_prediction') and col not in ['File Name_prediction', 'Species_prediction', 'Dataset_prediction']]

df_original_drugs_general = df_merged_general[original_cols_general].copy()
df_predictions_drugs_general = df_merged_general[prediction_cols_general].copy()

df_original_drugs_general.columns = [col.replace('_original', '') for col in df_original_drugs_general.columns]
df_predictions_drugs_general.columns = [col.replace('_prediction', '') for col in df_predictions_drugs_general.columns]

df_original_drugs_general = df_original_drugs_general.replace({'S': 1, 'R': 0, 'I': 1})
df_predictions_drugs_general = df_predictions_drugs_general.replace({'S': 1, 'R': 0, 'I': 1})

df_original_drugs_general = df_original_drugs_general.fillna(1)
df_predictions_drugs_general = df_predictions_drugs_general.fillna(1)

df_original_drugs_general = df_original_drugs_general.astype(int)
df_predictions_drugs_general = df_predictions_drugs_general.astype(int)

display(df_original_drugs_general.head())
display(df_predictions_drugs_general.head())
```

## Calculate metrics per drug (general model)

### Subtask:
Iterate through the drug columns and calculate the evaluation metrics (Balanced Accuracy, Precision, Recall, F1, Fβ, ROC AUC, Jaccard, MCC) for the `General Model` dataset.


**Reasoning**:
Iterate through the drug columns and calculate the evaluation metrics for the general model dataset.



```python
balanced_accuracy_scores_general = []
precision_scores_general = []
recall_scores_general = []
f1_scores_general = []
fbeta_scores_general = []
roc_auc_scores_general = []
jaccard_scores_general = []
mcc_scores_general = []

for drug_col in df_original_drugs_general.columns:
    original_labels = df_original_drugs_general[drug_col]
    prediction_labels = df_predictions_drugs_general[drug_col]

    balanced_accuracy_scores_general.append(balanced_accuracy_score(original_labels, prediction_labels))
    precision_scores_general.append(precision_score(original_labels, prediction_labels, pos_label=0, zero_division=0))
    recall_scores_general.append(recall_score(original_labels, prediction_labels, pos_label=0, zero_division=0))
    f1_scores_general.append(f1_score(original_labels, prediction_labels, pos_label=0, zero_division=0))
    fbeta_scores_general.append(fbeta_score(original_labels, prediction_labels, beta=0.5, pos_label=0, zero_division=0))

    if len(np.unique(original_labels)) > 1:
        roc_auc_scores_general.append(roc_auc_score(original_labels, prediction_labels))
    else:
        roc_auc_scores_general.append(np.nan)

    jaccard_scores_general.append(jaccard_score(original_labels, prediction_labels, pos_label=0, zero_division=0))
    mcc_scores_general.append(matthews_corrcoef(original_labels, prediction_labels))
```

## Create results dataframe (general model)

Create a DataFrame to store the calculated metrics for each drug for the `General Model` dataset.


**Reasoning**:
Create a DataFrame to store the calculated metrics for each drug for the General Model dataset as instructed and display its head.



```python
general_metrics_data = {
    'Drug': df_original_drugs_general.columns.tolist(),
    'Balanced Accuracy': balanced_accuracy_scores_general,
    'Precision': precision_scores_general,
    'Recall': recall_scores_general,
    'F1 Score': f1_scores_general,
    'Fbeta Score (beta=0.5)': fbeta_scores_general,
    'ROC AUC': roc_auc_scores_general,
    'Jaccard Score': jaccard_scores_general,
    'MCC': mcc_scores_general
}

df_results_general = pd.DataFrame(general_metrics_data)
display(df_results_general.head())
```

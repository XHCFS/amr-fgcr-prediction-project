```python
import os
import subprocess
import pandas as pd
from io import StringIO
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm.auto import tqdm

def split_drug_classes(drug_str):
    if pd.isna(drug_str):  # handle NaN values
        return []
    return [cls.strip() for cls in drug_str.split(';') if cls.strip()]

def process_sample(idx, row, root_dir, abricate_db, abricate_cmd, drug_class_mapping, drugs):
    """
    Helper function to process each sample in parallel.
    Returns (idx, classes_list).
    """
    fasta = os.path.join(root_dir,
                         row['Species'],
                         row['Dataset'],
                         row['File Name'])
    classes = []

    try:
        proc = subprocess.run(
            [abricate_cmd, '--db', abricate_db, fasta],
            capture_output=True, text=True, check=True
        )
        out_txt = proc.stdout
    except subprocess.CalledProcessError as e:
        # Warn on failure, but avoid per-sample debug output
        print(f"[WARNING] Abricate failed idx={idx} on {fasta!r}: {e.stderr or e.stdout}")
    else:
        if out_txt.strip() and not (
            out_txt.lstrip().startswith('#') and len(out_txt.splitlines()) <= 1
        ):
            df_abr = pd.read_csv(
                StringIO(out_txt), sep='\t', comment='#', header=None
            )
            df_abr.columns = [
                'genome_path','contig','start','end','strand','gene','query_hsp',
                'alignment_bar','errors','identity','coverage','db','reference_hit',
                'description','class'
            ]
            classes = (
                df_abr['class']
                .dropna()
                .str.lower()
                .str.split(';')
                .explode()
                .str.strip()
                .unique()
                .tolist()
            )

    return idx, classes


def generate_table_with_labels(master_table,
                               root_dir,
                               abricate_db='card',
                               drug_class_mapping=None,
                               abricate_cmd='abricate',
                               output_csv=None,
                               n_workers=None):
    """
    Parallel abricate + R/S mapping with a live tqdm bar and CSV flush every 50 samples.
    Resumes if output_csv exists by skipping processed samples based on File Name, Species, Dataset.
    Outputs CSV with columns: File Name, Species, Dataset, <drug cols>, raw_classes.
    """
    if drug_class_mapping is None:
        raise ValueError("Provide drug_class_mapping")

    # Meta and drug columns
    meta = ['File Name','Species','Dataset']
    drugs = [c for c in master_table.columns if c not in ['Species','Dataset','File Name']]

    # Build header
    header = meta + drugs + ['raw_classes']

    # Determine resume state
    processed_keys = set()
    if output_csv and os.path.exists(output_csv):
        try:
            df_done = pd.read_csv(output_csv, usecols=meta)
            processed_keys = set(zip(
                df_done['File Name'], df_done['Species'], df_done['Dataset']
            ))
            print(f"Resuming: {len(processed_keys)} samples already in {output_csv}")
        except Exception:
            print("Could not read existing CSV, starting fresh.")
    elif output_csv:
        # write header if CSV not exists
        pd.DataFrame(columns=header).to_csv(output_csv, index=False)
        print(f"Created new CSV {output_csv} with header")

    # Prepare output DataFrame
    df_out = master_table.copy()
    df_out[drugs] = ""

    # Filter to only unprocessed
    remaining = []
    for idx, row in master_table.iterrows():
        key = (row['File Name'], row['Species'], row['Dataset'])
        if key not in processed_keys:
            remaining.append(idx)
    print(f"{len(remaining)}/{len(master_table)} samples to process")

    # Progress bar
    pbar = tqdm(total=len(remaining), desc="Processing samples")

    buffer = []
    flush_interval = 20

    with ThreadPoolExecutor(max_workers=n_workers) as executor:
        futures = {executor.submit(
            process_sample,
            idx,
            master_table.loc[idx],
            root_dir,
            abricate_db,
            abricate_cmd,
            drug_class_mapping,
            drugs
        ): idx for idx in remaining}

        for future in as_completed(futures):
            idx = futures[future]
            idx_ret, classes = future.result()
            assert idx_ret == idx

            # Map to R/S and prepare row_data
            row = master_table.loc[idx]
            row_data = {k: row[k] for k in meta}
            for drug in drugs:
                want = drug_class_mapping.get(drug, '').lower()
                rs = 'R' if want in classes else 'S'
                df_out.at[idx, drug] = rs
                row_data[drug] = rs
            row_data['raw_classes'] = ';'.join(classes)

            buffer.append(row_data)

            # Flush buffer every flush_interval
            if output_csv and len(buffer) >= flush_interval:
                pd.DataFrame(buffer, columns=header).to_csv(
                    output_csv, mode='a', header=False, index=False
                )
                buffer.clear()

            pbar.update(1)

    pbar.close()

    # Final flush
    if output_csv and buffer:
        pd.DataFrame(buffer, columns=header).to_csv(
            output_csv, mode='a', header=False, index=False
        )
        buffer.clear()

    # Save full table if desired
    if output_csv:
        full_csv = output_csv.replace('.csv', '_fullRS.csv')
        df_out.to_csv(full_csv, index=False)
        print(f"Written full R/S table to {full_csv}")

    return df_out
```

```python
import os
import subprocess
import pandas as pd
from io import StringIO
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm.auto import tqdm


def split_drug_classes(drug_str):
    if pd.isna(drug_str):  # handle NaN values
        return []
    return [cls.strip() for cls in drug_str.split(';') if cls.strip()]


def process_sample(idx, row, root_dir, abricate_db, abricate_cmd):
    """
    Helper for fast version: runs abricate on a single sample.
    Returns (idx, list_of_classes).
    """
    fasta = os.path.join(root_dir, row['Species'], row['Dataset'], row['File Name'])
    classes = []
    try:
        proc = subprocess.run(
            [abricate_cmd, '--db', abricate_db, fasta],
            capture_output=True, text=True, check=True
        )
        out_txt = proc.stdout
    except subprocess.CalledProcessError as e:
        print(f"[WARNING] Abricate failed idx={idx} on {fasta!r}: {e.stderr or e.stdout}")
        return idx, []

    if out_txt.strip() and not (out_txt.lstrip().startswith('#') and len(out_txt.splitlines()) <= 1):
        df = pd.read_csv(
            StringIO(out_txt), sep='\t', comment='#', header=None
        )
        df.columns = [
            'genome_path','contig','start','end','strand','gene','query_hsp',
            'alignment_bar','errors','identity','coverage','db','reference_hit',
            'description','class'
        ]
        classes = (
            df['class']
            .dropna()
            .str.lower()
            .str.split(';')
            .explode()
            .str.strip()
            .unique()
            .tolist()
        )
    return idx, classes


def generate_table_with_labels_fast(master_table,
                                    root_dir,
                                    abricate_db='card',
                                    drug_class_mapping=None,
                                    abricate_cmd='abricate',
                                    output_csv=None,
                                    n_workers=None,
                                    flush_interval=20):
    """
    Fast version: uses ProcessPoolExecutor to parallelize abricate calls.
    """
    if drug_class_mapping is None:
        raise ValueError("Provide drug_class_mapping")

    meta = ['File Name','Species','Dataset']
    drugs = [c for c in master_table.columns if c not in meta]
    header = meta + drugs + ['raw_classes']

    # Resume logic
    processed = set()
    if output_csv and os.path.exists(output_csv):
        df_done = pd.read_csv(output_csv, usecols=meta, dtype=str)
        processed = set(zip(df_done['File Name'], df_done['Species'], df_done['Dataset']))
    elif output_csv:
        pd.DataFrame(columns=header).to_csv(output_csv, index=False)

    df_out = master_table.copy()
    df_out[drugs] = ""
    remaining = [idx for idx, row in master_table.iterrows()
                 if (row['File Name'], row['Species'], row['Dataset']) not in processed]
    print(f"{len(remaining)}/{len(master_table)} samples to process (fast)")

    pbar = tqdm(total=len(remaining), desc="Fast processing")
    buffer = []

    if n_workers is None:
        n_workers = os.cpu_count() or 4

    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        futures = {executor.submit(
            process_sample, idx, master_table.loc[idx], root_dir, abricate_db, abricate_cmd
        ): idx for idx in remaining}

        for future in as_completed(futures):
            idx = futures[future]
            _, classes = future.result()

            row = master_table.loc[idx]
            row_data = {k: row[k] for k in meta}
            for drug in drugs:
                want = drug_class_mapping.get(drug, '').lower()
                rs = 'R' if want in classes else 'S'
                df_out.at[idx, drug] = rs
                row_data[drug] = rs
            row_data['raw_classes'] = ';'.join(classes)

            buffer.append(row_data)
            if output_csv and len(buffer) >= flush_interval:
                pd.DataFrame(buffer, columns=header).to_csv(
                    output_csv, mode='a', header=False, index=False
                )
                buffer.clear()
            pbar.update(1)

    pbar.close()
    if output_csv and buffer:
        pd.DataFrame(buffer, columns=header).to_csv(
            output_csv, mode='a', header=False, index=False
        )
    # Save full table
    if output_csv:
        full_csv = output_csv.replace('.csv', '_fullRS.csv')
        df_out.to_csv(full_csv, index=False)
        print(f"Written full R/S table to {full_csv}")
    return df_out
```

```python
import os
import subprocess
import pandas as pd
from io import StringIO
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm.auto import tqdm
import math
import time
import shutil # Needed for file copying and directory removal

# Keep this helper function as it's useful for parsing class strings
def split_drug_classes(drug_str):
    if pd.isna(drug_str):  # handle NaN values
        return []
    return [cls.strip() for cls in drug_str.split(';') if cls.strip()]


# Modified process_batch: Doesn't need root_dir, works with provided (local) paths
def process_batch(batch_info, abricate_db, abricate_cmd):
    """
    Processes a BATCH of samples using a single Abricate call on LOCAL file paths.

    Args:
        batch_info (list): A list of tuples, where each tuple is
                           (idx, local_fasta_path, species, dataset).
        abricate_db (str): Abricate database name.
        abricate_cmd (str): Path to the abricate executable.

    Returns:
        dict: A dictionary mapping original index (idx) to a list of
              identified raw AMR classes (lowercased, unique).
              Returns None if the entire batch fails.
    """
    results = {}
    # Paths are now local paths provided by the main function
    local_fasta_files = [item[1] for item in batch_info]
    # Map the local path back to the original index
    idx_map = {item[1]: item[0] for item in batch_info}

    # Initialize results for all files in the batch
    for idx, _, _, _ in batch_info:
        results[idx] = []

    if not local_fasta_files:
        return {}

    # Construct the command using the local file paths
    command = [abricate_cmd, '--db', abricate_db] + local_fasta_files

    try:
        proc = subprocess.run(
            command,
            capture_output=True, text=True, check=True, timeout=600
        )
        out_txt = proc.stdout
    except subprocess.CalledProcessError as e:
        # Use the first local path for context in error message
        context_path = local_fasta_files[0] if local_fasta_files else "N/A"
        print(f"[ERROR] Abricate batch failed (local files starting with {os.path.basename(context_path)}). Error: {e.stderr or e.stdout}")
        return results # Return empty results for the batch
    except subprocess.TimeoutExpired:
        context_path = local_fasta_files[0] if local_fasta_files else "N/A"
        print(f"[ERROR] Abricate batch timed out (local files starting with {os.path.basename(context_path)}).")
        return results # Return empty results

    # Process the combined output
    if out_txt.strip() and not (
        out_txt.lstrip().startswith('#') and len(out_txt.splitlines()) <= 1
    ):
        try:
            df_abr_batch = pd.read_csv(
                StringIO(out_txt), sep='\t', comment='#', header=None
            )
            df_abr_batch.columns = [
                'FILE', 'SEQUENCE', 'START', 'END', 'STRAND', 'GENE', 'COVERAGE',
                'COVERAGE_MAP', 'GAPS', '%COVERAGE', '%IDENTITY', 'DATABASE',
                'ACCESSION', 'PRODUCT', 'RESISTANCE'
            ]

            # Group by the LOCAL file path ('FILE' column) output by abricate
            for local_file_path, group in df_abr_batch.groupby('FILE'):
                if local_file_path in idx_map:
                    idx = idx_map[local_file_path]
                    classes = (
                        group['RESISTANCE']
                        .dropna()
                        .str.lower()
                        .str.split(';')
                        .explode()
                        .str.strip()
                        .unique()
                        .tolist()
                    )
                    results[idx] = classes
                else:
                     # This might happen if abricate outputs paths differently than expected
                     print(f"[WARNING] Abricate output file '{local_file_path}' not found in input batch mapping.")

        except pd.errors.EmptyDataError:
            pass # No results found
        except Exception as e:
            print(f"[ERROR] Failed to parse Abricate output for batch. Error: {e}")
            # Return potentially partial results
            return results

    return results


# Assuming process_batch is defined elsewhere and runs Abricate on the batch
# def process_batch(batch_info, abricate_db, abricate_cmd):
#     pass

def generate_table_with_labels_batched_local_copy(master_table,
                                                 root_dir,
                                                 abricate_db='card',
                                                 drug_class_mapping=None,
                                                 abricate_cmd='abricate',
                                                 output_csv=None,
                                                 n_workers=2,
                                                 batch_size=1,
                                                 reverse_order=True):
    """
    Parallel abricate processing with local copies. Copies files to a buffer,
    runs abricate in batches, and maps R/S labels. Buffer is cleared only after
    all batches are processed. Resumes if output_csv exists.
    """
    start_time = time.time()
    if drug_class_mapping is None:
        raise ValueError("Provide drug_class_mapping")
    if not os.path.isdir(root_dir):
        raise FileNotFoundError(f"Root directory not found: {root_dir}")

    # Local Buffer Setup
    buffer_dir = "/content/abricate_buffer"
    print(f"Using local buffer directory: {buffer_dir}")
    if os.path.exists(buffer_dir):
        print("Cleaning up existing buffer directory...")
        shutil.rmtree(buffer_dir)
    os.makedirs(buffer_dir, exist_ok=True)

    # Meta and drug columns
    meta = ['File Name', 'Species', 'Dataset']
    drugs = [c for c in master_table.columns if c not in ['Species', 'Dataset', 'File Name']]
    header = meta + drugs + ['raw_classes']

    # Resume Logic
    processed_keys = set()
    initial_file_exists = False
    if output_csv and os.path.exists(output_csv):
        initial_file_exists = True
        try:
            df_done = pd.read_csv(output_csv, usecols=meta)
            processed_keys = set(zip(df_done['File Name'], df_done['Species'], df_done['Dataset']))
            print(f"Resuming: {len(processed_keys)} samples already processed based on {output_csv}")
        except Exception as e:
            print(f"Warning: Could not read existing CSV '{output_csv}' to resume: {e}. Starting fresh or appending.")
            processed_keys = set()
            initial_file_exists = False
    elif output_csv:
        print(f"Output CSV '{output_csv}' not found, will create a new one.")

    # Prepare Data
    tasks_to_run_info = []
    indices_to_process = []
    df_out = master_table.copy()
    for drug in drugs:
        if drug not in df_out.columns:
            df_out[drug] = 'S'
        else:
            df_out[drug] = df_out[drug].astype(str)
    if 'raw_classes' not in df_out.columns:
        df_out['raw_classes'] = ''
    else:
        df_out['raw_classes'] = df_out['raw_classes'].astype(str)

    print("Identifying samples to process...")
    duplicate_filenames_detected = False
    filenames_seen = set()
    tasks_to_run_map = {}

    for idx, row in tqdm(master_table.iterrows(), total=len(master_table), desc="Scanning table"):
        key = (row['File Name'], row['Species'], row['Dataset'])
        if key not in processed_keys:
            drive_path = os.path.join(root_dir, row['Species'], row['Dataset'], row['File Name'])
            base_name = os.path.basename(drive_path)
            if base_name in filenames_seen:
                duplicate_filenames_detected = True
            filenames_seen.add(base_name)

            if not os.path.exists(drive_path):
                print(f"[WARNING] Original FASTA file not found for index {idx}: {drive_path}. Skipping.")
                df_out.loc[idx, drugs] = 'S'
                df_out.loc[idx, 'raw_classes'] = 'FILE_NOT_FOUND'
                processed_keys.add(key)
                continue

            task_data = (idx, drive_path, row['Species'], row['Dataset'])
            tasks_to_run_info.append(task_data)
            indices_to_process.append(idx)
            tasks_to_run_map[idx] = task_data

    if duplicate_filenames_detected:
        print("[INFO] Duplicate base filenames detected. Using index-prefixed names in buffer.")

    if not tasks_to_run_info:
        print("No new samples to process.")
        if output_csv and initial_file_exists:
            print(f"Loading final results from {output_csv}")
            try:
                return pd.read_csv(output_csv)
            except Exception as e:
                print(f"[ERROR] Failed to load existing CSV {output_csv}: {e}")
                return df_out
        return df_out

    print(f"{len(tasks_to_run_info)}/{len(master_table)} samples to process.")


    if reverse_order:
        print("‚ö° Running in REVERSE order: processing unfinished samples from the end first.")
        tasks_to_run_info = list(reversed(tasks_to_run_info))


    if output_csv and not initial_file_exists:
        pd.DataFrame(columns=header).to_csv(output_csv, index=False)
        print(f"Created new CSV {output_csv} with header")

    # Execute in Parallel with Local Copy
    results_buffer = []
    flush_interval_batches = 5
    batches_processed_since_flush = 0
    num_workers = min(os.cpu_count() or 1, (len(tasks_to_run_info) // batch_size) + 1) if n_workers is None else n_workers
    total_batches = math.ceil(len(tasks_to_run_info) / batch_size)
    print(f"Processing {total_batches} batches with {num_workers} workers (batch size: {batch_size})...")

    try:
        with ThreadPoolExecutor(max_workers=num_workers) as executor:
            futures = []
            for i in tqdm(range(0, len(tasks_to_run_info), batch_size), desc="Submitting batches"):
                current_batch_drive_info = tasks_to_run_info[i:i + batch_size]
                current_batch_local_info = []
                copied_successfully_count = 0

                # Copy files for the current batch (buffer is NOT cleared here)
                for idx, drive_path, species, dataset in current_batch_drive_info:
                    try:
                        base_name = os.path.basename(drive_path)
                        local_base_name = f"{idx}_{base_name}" if duplicate_filenames_detected else base_name
                        local_path = os.path.join(buffer_dir, local_base_name)
                        shutil.copy2(drive_path, local_path)
                        if os.path.exists(local_path) and os.path.getsize(local_path) > 0:
                            current_batch_local_info.append((idx, local_path, species, dataset))
                            copied_successfully_count += 1
                        else:
                            print(f"[WARNING] Copied file missing or empty: {local_path}. Original: {drive_path}. Skipping index {idx}.")
                            df_out.loc[idx, drugs] = 'S'
                            df_out.loc[idx, 'raw_classes'] = 'FILE_COPY_FAILED_OR_MISSING'
                    except FileNotFoundError:
                        print(f"[WARNING] Original file not found during copy: {drive_path}. Skipping index {idx}.")
                        df_out.loc[idx, drugs] = 'S'
                        df_out.loc[idx, 'raw_classes'] = 'FILE_NOT_FOUND_DURING_COPY'
                    except Exception as e:
                        print(f"[ERROR] Failed to copy {drive_path} to buffer: {e}. Skipping index {idx}.")
                        df_out.loc[idx, drugs] = 'S'
                        df_out.loc[idx, 'raw_classes'] = 'FILE_COPY_ERROR'

                if current_batch_local_info:
                    future = executor.submit(process_batch, current_batch_local_info, abricate_db, abricate_cmd)
                    futures.append(future)

            # Process results
            pbar = tqdm(total=len(tasks_to_run_info), desc="Processing samples")
            for future in as_completed(futures):
                try:
                    batch_results = future.result()
                    if batch_results is None:
                        print(f"[ERROR] Batch failed catastrophically (returned None).")
                        pbar.update(batch_size)
                        continue

                    rows_to_append = []
                    processed_indices_count = 0
                    for idx, classes in batch_results.items():
                        processed_indices_count += 1
                        original_info = tasks_to_run_map.get(idx)
                        if original_info:
                            _, original_drive_path, species, dataset = original_info
                            filename = os.path.basename(original_drive_path)
                            row_data = {'File Name': filename, 'Species': species, 'Dataset': dataset}
                            for drug in drugs:
                                required_class = drug_class_mapping.get(drug, '').lower()
                                rs = 'R' if required_class and required_class in classes else 'S'
                                df_out.loc[idx, drug] = rs
                                row_data[drug] = rs
                            raw_classes_str = ';'.join(sorted(classes)) if classes else ''
                            df_out.loc[idx, 'raw_classes'] = raw_classes_str
                            row_data['raw_classes'] = raw_classes_str
                            rows_to_append.append(row_data)
                        else:
                            print(f"[ERROR] Metadata missing for index {idx}.")
                            df_out.loc[idx, drugs] = 'S'
                            df_out.loc[idx, 'raw_classes'] = 'METADATA_ERROR_POST_ABRICATE'
                            row_data = {
                                'File Name': f"UNKNOWN_FILE_FOR_IDX_{idx}",
                                'Species': 'Unknown',
                                'Dataset': 'Unknown',
                                'raw_classes': 'METADATA_ERROR_POST_ABRICATE'
                            }
                            for drug in drugs:
                                row_data[drug] = 'S'
                            rows_to_append.append(row_data)

                    results_buffer.extend(rows_to_append)
                    batches_processed_since_flush += 1
                    if output_csv and batches_processed_since_flush >= flush_interval_batches:
                        print(f"Flushing {len(results_buffer)} results to CSV...")
                        pd.DataFrame(results_buffer, columns=header).to_csv(
                            output_csv, mode='a', header=False, index=False
                        )
                        results_buffer.clear()
                        batches_processed_since_flush = 0
                    pbar.update(processed_indices_count)
                except Exception as e:
                    print(f"[ERROR] Failed processing batch: {e}")
                    pbar.update(batch_size)

    finally:
        print(f"Cleaning up buffer directory: {buffer_dir}")
        if os.path.exists(buffer_dir):
            try:
                shutil.rmtree(buffer_dir)
            except Exception as e:
                print(f"[ERROR] Failed to remove buffer directory {buffer_dir}: {e}")

    # Final Flush and Save
    if output_csv and results_buffer:
        print(f"Final flush: Writing {len(results_buffer)} results to CSV...")
        pd.DataFrame(results_buffer, columns=header).to_csv(
            output_csv, mode='a', header=False, index=False
        )

    if output_csv:
        try:
            print(f"Saving final table to {output_csv}...")
            ordered_cols = header + [col for col in df_out.columns if col not in header]
            df_out[ordered_cols].to_csv(output_csv, index=False)
            print("Save complete.")
        except Exception as e:
            print(f"[ERROR] Failed to save CSV to {output_csv}: {e}")

    end_time = time.time()
    print(f"Processing finished in {end_time - start_time:.2f} seconds.")
    return df_out
```
